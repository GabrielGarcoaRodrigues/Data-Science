#Data Frame

```python
dados = [[1, 2, 3],
        [4, 5, 6],
        [7, 8, 9]]

dados
```

```python
df1 = pd.DataFrame(data = dados)   #Criando um data frame e passando os dados
df1
```

```python
index = ["Linha " + str(i+1) for i in range(3)]         #Criando o index e a coluna para o df
columns = ["Coluna " + str(i+1) for i in range(3)]
```

```python
df1 = pd.DataFrame(data = dados, index = index, columns = columns)  #Passando o index e coluna para o df
df1
```

```python
df1[df1 > 0] = 'A'     #Elementos maiores que 0, troca para A
df1
```

```python
df2 = pd.DataFrame(data = dados, index = index, columns = columns) #Criando df igual
df3 = pd.DataFrame(data = dados, index = index, columns = columns)

df2[df2 > 0] = 'B'    #Trocando tudo por B
df3[df3 > 0] = 'C'    #Trocando tudo por C
```

```python
df4 = pd.concat([df1, df2, df3])      #Concatenando os dfs
df4
```

```python
df4 = pd.concat([df1, df2, df3], axis = 1)  #Concatenando por linha
df4
```
